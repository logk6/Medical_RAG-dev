```python
import os
import langchain_community
import json
import pandas as pd
import torch
from tqdm import tqdm
from pathlib import Path
from functools import partial
from bioBERT_encoder import bioBERTEncoder
from medCPT_encoder import medCPTArticleEncoder

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder = bioBERTEncoder()
print(f"Using {device}.")
```

```python
source_file = Path('../../data/embed_data/source/text_chunked.jsonl')
target_directory = Path('../../data/embed_data/target')
target_file = target_directory / 'text_chunked.jsonl'
target_directory.mkdir(parents=True, exist_ok=True)

encoder.embed_file(target_file, source_file)
# for file_name in tqdm(os.listdir(source_directory)):
#     if file_name.endswith('.jsonl'):

# Tạo một tệp mới trong thư mục đích
with open(target_file, 'w', encoding='utf-8') as target:
    with open(source_file, 'r', encoding='utf-8') as source:
        for line in source:
            # Mỗi dòng là một đối tượng JSON
            item = json.loads(line)
            embedded_item = encoder.pdf_encode(item)
            # Ghi đối tượng đã embed vào tệp đích
            target.write(json.dumps(embedded_item) + '\n')
    print(f"{target_file} has been successfully written to data/embed_data/target.")

print("Tất cả các tập tin đã được xử lý và lưu.")
```
